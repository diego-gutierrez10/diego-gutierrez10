```python
#We import the libraries needed and create a dataframe from the given CSV

import pandas as pd
import numpy as np
pharma_data = pd.read_csv('https://raw.githubusercontent.com/dphi-official/Datasets/master/pharma_data/Training_set_begs.csv')
```

```python
#We portray the head of the Dataframe to get an idea of how the values are like

pharma_data.head()
```

```python
#For an easier processing of the model we drop the rows with empty values (NaN)

#pharma_data.isnull().sum()
pharma_data = pharma_data.dropna(axis=0)
```

```python
#We portray the description of the Dataframe to get an idea of the values that we are working with

#pharma_data.columns
pharma_data.describe()
```

```python
#We make dummies (aka 'One-hot econding') for the non-numerical values

pharma_data = pd.get_dummies(pharma_data, prefix_sep='_', drop_first=True)
pharma_data.columns
```

```python
#We separate our "X" (What the model will learn from) and "Y"(The prediction target)

X = pharma_data.drop('Survived_1_year', axis = 1)
y = pharma_data.Survived_1_year
```

# **ML Model making:**

```python
#We import the machine learning library (scikit-learn), and split our "X" and "y" by 2/3 for the model training and 1/3 for the model testing

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=66)

x_train.head()
y_train.head()
```

```python
#We check that there's no missing values left

x_train.isnull().sum()
```

```python
#As the model target is a 'Yes or No' answer, we will use DecisionTreeClassifier: We import it and fit the model with the training X and y
from sklearn.tree import DecisionTreeClassifier

pharma_model = DecisionTreeClassifier(random_state=66, max_depth=7, min_samples_split=4)
pharma_model.fit(x_train, y_train)
print(pharma_model.predict(x_train.head(50)))
print(y_train.head(50))
```

## We have now finshed making the model that allows us to predict if a patient will survive or not.

## Finally, we need to test it on the unseen third of the data or "x_test"

```python
from sklearn.metrics import accuracy_score
print('Training accuracy:',accuracy_score(y_train, pharma_model.predict(x_train)))
print('Test accuracy:', accuracy_score(y_test, pharma_model.predict(x_test)))
```

```python
#We make an array for the predictions of our model over x_test

y_true = pharma_model.predict(x_test)
```

```python
#We check what we get on the f1_score

from sklearn.metrics import f1_score

print('The f1_score is',f1_score(y_test, y_true))
```

## **Now that we have finished making and evaluating or ML model, we will make the final prediction for the submission on the challenge**

```python
#We import the CSV and make a dataframe for it

test_data = pd.read_csv('https://raw.githubusercontent.com/dphi-official/Datasets/master/pharma_data/Testing_set_begs.csv')
```

```python
#We take a look to the values present in the test_data Dataframe

test_data.head()
```

```python
#We check that tere's no missing values in the test_data Dataframe


test_data.isnull().sum()
```

```python
#We make dummies as the ones in the pharma_data Dataframe

test_data = pd.get_dummies(test_data, prefix_sep='_', drop_first=True)
test_data.columns
```

```python
#We make another dataframe (prediction_CSV) for the pharma_model predictions based on test_data

prediction_CSV = pd.DataFrame((pharma_model.predict(test_data)), columns=['prediction'])
prediction_CSV.head()
```
